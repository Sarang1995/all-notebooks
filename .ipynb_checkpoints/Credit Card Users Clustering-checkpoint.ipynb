## silhouette_score

```python
s_score = []

for i in range(2,16):
    kmean = KMeans(n_clusters=i, n_init=10, random_state=42)
    kmean.fit(df)
    score = silhouette_score(df, kmean.labels_, metric='euclidean', sample_size=None)
    s_score.append(score)
    
s_score = pd.Series(s_score, index=range(2,16))

s_score
```

## DBSCAN

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)

dbscan.fit(df)
```

```python
def tune_dbscan(data):
    
    result = []
    
    eps_values = np.arange(0.1, 2, 0.1)
    min_sample_values = np.arange(2, 10, 1)
    
    ## For loop to train model on different eps and min_samples
    
    for eps in eps_values:
        for samples in min_sample_values:
            dbscan = DBSCAN(eps=eps, min_samples=samples)
            dbscan.fit(data)
            labels = dbscan.labels_
            
            ## Number of clusters (excluding noise)
            n_clusters = len(set(dbscan.labels_)) - (1 if -1 in labels else 0)
            
            ## Number of noise
            n_noise = list(dbscan.labels_).count(-1)
            
            ## calculate silhouette score (excluding noise point)  
            if n_clusters > 1: #silhouette score need minimum of 2 clusters
                score = silhouette_score(data, labels, metric='euclidean', sample_size=None)
            else:      
                score = None
                
            result.append([eps, samples, n_clusters, n_noise, score])
            
    ## Put the result in dataframe        
    result_df = pd.DataFrame(result, columns=['eps','min_samples', 'n_clusters','n_noise','silhouette score'] )
    
    return result_df
```

```python
dbscan_result = tune_dbscan(df)
dbscan_result.head()
```

```python
dbscan_result.sort_values("silhouette score", ascending=False).head()
```

```python
(dbscan_result.sort_values("silhouette score", ascending=False)
 .groupby('n_clusters').head(1)).head()
```

```python
def print_metrics(model, data):
    print(model)
    print(Counter(model.labels_))
    print(silhouette_score(data, model.labels_))
```

## Reccommendations for Marketing Startegy

Based on the results of the profiling above, here are some suggestions for marketing strategies:


* **Cluster 1** users are not active credit card users. Threfore, the banks can give them special offers and encourage them to use credit card by telling them benefits of credit cards. Banks can create **Engagement Campaigns**, **Educational Content**, **Exclusive Offers**, etc.

* **Cluster 2** users has high balance, moderate credit limit and love to pay cash in advance. By observing these charecteristics banks can encourage them to purchase using credit card by telling them benifits of credit card vs paying cash.

* Users form **Cluster 3** can be the main target for the bank. This is because these users are highly active in using credit cards. These users has high tenure, high credit limits and they purchase a lot. Banks can make a profit by targeting these users. Looking at these users profile we can say that these users have high perchasing power. Banks can cross sell some other products and also offer them exclusive VIP rewards for their loyalty.
